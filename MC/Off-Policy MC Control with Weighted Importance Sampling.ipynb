```python
def create_random_policy(nA):
    """
    Creates a random policy function.
    
    Args:
        nA: Number of actions in the environment.
    
    Returns:
        A function that takes an observation as input and returns a vector
        of action probabilities
    """
    A = np.ones(nA, dtype=float) / nA
    def policy_fn(observation):
        return A
    return policy_fn
```

```python
def create_greedy_policy(Q, nA):
    """
    Creates a greedy policy based on Q values.
    
    Args:
        Q: A dictionary that maps from state -> action values
        
    Returns:
        A function that takes an observation as input and returns a vector
        of action probabilities.
    """
    
    def policy_fn(observation):
        A=np.zeros(nA)
        best_action=np.argmax(Q[observation])
        A[best_action]=1
    return policy_fn
```

```python
def mc_control_importance_sampling(env, num_episodes, behavior_policy, discount_factor=1.0):
    """
    Monte Carlo Control Off-Policy Control using Weighted Importance Sampling.
    Finds an optimal greedy policy.
    
    Args:
        env: OpenAI gym environment.
        num_episodes: Number of episodes to sample.
        behavior_policy: The behavior to follow while generating episodes.
            A function that given an observation returns a vector of probabilities for each action.
        discount_factor: Gamma discount factor.
    
    Returns:
        A tuple (Q, policy).
        Q is a dictionary mapping state -> action values.
        policy is a function that takes an observation as an argument and returns
        action probabilities. This is the optimal greedy policy.
    """
    
    # The final action-value function.
    # A dictionary that maps state -> action values
    Q = defaultdict(lambda: np.zeros(env.action_space.n))
    N = defaultdict(float)
    D = defaultdict(float)
    
    
    # Our greedily policy we want to learn
    target_policy = create_greedy_policy(Q, env.action_space.n)

    # Implement this!
    #select behavior policy and use it to generate episodes
    for i_episode in range(num_episodes):
        
        #print episode num
        if i_episode % 1000 == 0:
            print("\rEpisode {}/{}.".format(i_episode, num_episodes))
            sys.stdout.flush()
        
        
        episode = []
        state = env.reset()
        record=True
        done=False
        for t in range(100):
            behavior_prob = behavior_policy(state)
            target_prob = target_policy(state)
            behavior_action = np.random.choice(np.arange(env.action_space.n), p=behavior_prob)
            target_action = np.random.choice(np.arange(env.action_space.n), p=target_prob)
            next_state, reward, done, _ = env.step(behavior_action)
            
            if behavior_action!=target_action:
                #only record after T<<time where 
                record=True
                
            if record:
                episode.append((state, behavior_action, reward))
            
            if done:
                break
            
            state = next_state
        
        for x in episode:
            state=x[0]
            action=x[1]
            sa_pair=(state, action)
            
            W=1
            for i in range(len(episode)):
                W/=behavior_policy(state)[action]
                
                #compute G
                G=0
                first=True
                for i in range(len(episode)):
                    if state==episode[i][0] and action==episode[i][1] and first:
                        first=False
                        first_idx=i
                    if not first:
                        n=i-first_idx
                        reward=episode[i][2]
                        G+=reward*(discount_factor**n)

            N[sa_pair]+=W*G
            D[sa_pair]+=W
            Q[state][action]=N[sa_pair]/D[sa_pair]
    
    
    
    return Q, target_policy
```

```python
random_policy = create_random_policy(env.action_space.n)
Q, policy = mc_control_importance_sampling(env, num_episodes=500000, behavior_policy=random_policy)
```
